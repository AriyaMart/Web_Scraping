```python
records=[]
for e in el[:]:
    model=e.select_one("div.model_overview_list_btn").text.strip()
    description = e.select_one("div.model_overview_list_txt").text.strip()
    starting_price=e.select_one("div.model_overview_list_txt_price").text.strip()
    price = float(re.sub(r'\D','', starting_price))
    #print(model, description, price)
    records.append([model, description, price])

records
```

```python
df=pd.DataFrame(records, columns=['model','description','starting_price'])
df
```
