```python
data = driver.page_source
type(data)
```

```python
import bs4
soup = bs4.BeautifulSoup(data)
type(soup)
```

```python
product_name = soup.find_all('div',{'class':'RfADt'})
len(product_name)
```

```python
product_list_name = []
for name in product_name:
    #print(name.text)
    #print('--'*30)
    product_list_name.append(name.text)
product_list_name
```

```python
product_price = soup.find_all('div',{'class':'aBrP0'})
product_price
```

```python
product_list_price = []
for price in product_price:
    product_list_price.append(price.text)
product_list_price
```

```python
place_list=soup.find_all('span',{'class':'oa6ri'})
place_list 
```

```python
place_store_list=[]
for place in place_list:
    #print(place.text)
    place_store_list.append(place.text)

place_store_list
```

```python
import pandas as pd
```

```python
df = pd.DataFrame([product_list_name, product_list_price, place_store_list])
df
```

```python
Lazada_data = df.transpose()
pd.set_option('display.max_colwidth', -1)
Lazada_data
```

```python
Lazada_data.columns=['title','price','place']
```
